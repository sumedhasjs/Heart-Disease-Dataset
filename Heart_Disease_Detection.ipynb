```python
import numpy as np
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
headings ={'age':"Age",'sex':"Sex",'cp':"Chest Pain Type",'trestbps':"Resting Blood Pressure",'chol':"Cholestrol",'fbs':'Fasting Blood Sugar','restecg':'Resting Electrocardiographic Results','thalach':'Maximum Heart Rate Achieved','exang':'Exercise Induced Angina','oldpeak':'ST Depression Induced','slope':'Slope of Peak ST Segment','ca':'Number of Major Blood Vessels','thal':'Type','target':'Target'}
heart.rename(columns=headings, inplace=True)
```

```python
heart.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
test_ratio=0.1
train_ratio=0.9
```

```python
Train,Test = train_test_split(heart,shuffle=True)
```

```python
from sklearn.svm import SVC
classifier = SVC(C=1000, kernel = 'rbf', random_state = 0)

classifier.fit(Train[["Age",'Sex','Chest Pain Type',	'Resting Blood Pressure',	'Cholestrol',	'Fasting Blood Sugar',	'Resting Electrocardiographic Results','Maximum Heart Rate Achieved','Exercise Induced Angina',	'ST Depression Induced',	'Slope of Peak ST Segment',	'Number of Major Blood Vessels',	'Type']],Train['Target'])
```

```python
y_pred_test = classifier.predict(Test[["Age",'Sex','Chest Pain Type',	'Resting Blood Pressure',	'Cholestrol',	'Fasting Blood Sugar',	'Resting Electrocardiographic Results','Maximum Heart Rate Achieved','Exercise Induced Angina',	'ST Depression Induced',	'Slope of Peak ST Segment',	'Number of Major Blood Vessels',	'Type']])
```

```python
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,classification_report
precision = precision_score(Test["Target"], y_pred_test)
accuracy = accuracy_score(Test["Target"], y_pred_test)
recall=recall_score(Test["Target"], y_pred_test)
f1=f1_score(Test["Target"], y_pred_test)
ClassificationReport=classification_report(Test["Target"], y_pred_test)
print("\nAccuracy: ",accuracy)
print("\nPrecision: ",precision)
print("\nRecall: ",recall)
print("\nF1 Score: ",f1)
print("\nClassification Report: \n")
print(ClassificationReport)
```
